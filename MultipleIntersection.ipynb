```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Import RSE25 data

```python
rse25 = pd.read_csv("data/RSE25.csv")
```

```python
rse25_upd = rse25[['Timestamp', 'CurrentIndication_MajorStreet', 'Duration_Indication_MajorStreet', 'CurrentIndication_MinorStreet', 'Duration_Indication_MinorStreet', 'RSE_Number']]
```

```python
timelist_25 = [datetime.fromtimestamp(l).strftime('%m-%d:%H') for l in rse25_upd['Timestamp'].tolist()]
```

```python
rse25_upd.insert(1, 'Time', timelist_25)
```

```python
rse25_upd.head()
```

```python
df_25_group = rse25_upd.groupby(['Time', 'CurrentIndication_MajorStreet'])
df_25_group_minor = rse25_upd.groupby(['Time', 'CurrentIndication_MinorStreet'])
```

```python
rse25_major_light_duration = []
for key1, key2 in df_25_group.groups.keys():
    if key2 == 1 or key2 == 3:
        time_list_red_major = list(df_25_group.get_group((key1, key2))['Duration_Indication_MajorStreet'])
        time_red = time_list_red_major[0]
        count_red = 0
        for i in time_list_red_major[1:]:
            if time_red > i:
                count_red = count_red + time_red
            time_red = i
        li = [key1, key2, int(count_red)]
        rse25_major_light_duration.append(li)
```

```python
rse25_major_light_duration
```

```python
red_25_major_list = []
green_25_major_light = []
for l in rse25_major_light_duration:
    if l[1] == 3:
        red_25_major_list.append(l[2])
    else:
        green_25_major_light.append(l[2])
```

```python
green_25_major_light
```

```python
rse25_minor_light_duration = []
for key1, key2 in df_25_group_minor.groups.keys():
    if key2 == 1 or key2 == 3:
        time_list_red_major = list(df_25_group_minor.get_group((key1, key2))['Duration_Indication_MinorStreet'])
        time_red = time_list_red_major[0]
        count_red = 0
        for i in time_list_red_major[1:]:
            if time_red > i:
                count_red = count_red + time_red
            time_red = i
        li = [key1, key2, int(count_red)]
        rse25_minor_light_duration.append(li)
```

Major Road Distribution

```python
rse25_major_light_count = []

for key1, key2 in df_25_group.groups.keys():
    if key2 == 1 or key2 == 3:
        time_list_red_major = list(df_25_group.get_group((key1, key2))['Duration_Indication_MajorStreet'])
        time_red = time_list_red_major[0]
        count_red = 0
        count_light = 0
        for i in time_list_red_major[1:]:
            if time_red > i:
                rse25_major_light_count.append([key2, int(time_red)])
            time_red = i
```

```python
one_25_count = []
three_25_count = []
for i in range(len(rse25_major_light_count)):
    if rse25_major_light_count[i][0] == 1:
        one_25_count.append(rse25_major_light_count[i][1])
    elif rse25_major_light_count[i][0] == 3:
        three_25_count.append(rse25_major_light_count[i][1])
```

```python
one_25_categories = [0]*8
three_25_categories = [0]*8

for num in one_25_count:
    if num in range(0, 16):
        one_25_categories[0] = one_25_categories[0] + 1
    elif num in range(16, 31):
        one_25_categories[1] = one_25_categories[1] + 1
    elif num in range(31, 45):
        one_25_categories[2] = one_25_categories[2] + 1
    elif num in range(45, 61):
        one_25_categories[3] = one_25_categories[3] + 1
    elif num in range(61, 76):
        one_25_categories[4] = one_25_categories[4] + 1
    elif num in range(76, 91):
        one_25_categories[5] = one_25_categories[5] + 1
    elif num in range(91, 106):
        one_25_categories[6] = one_25_categories[6] + 1
    elif num in range(106, 121):
        one_25_categories[7] = one_25_categories[7] + 1


for num in three_25_count:
    if num in range(0, 16):
        three_25_categories[0] = three_25_categories[0] + 1
    elif num in range(16, 31):
        three_25_categories[1] = three_25_categories[1] + 1
    elif num in range(31, 45):
        three_25_categories[2] = three_25_categories[2] + 1
    elif num in range(45, 61):
        three_25_categories[3] = three_25_categories[3] + 1
    elif num in range(61, 76):
        three_25_categories[4] = three_25_categories[4] + 1
    elif num in range(76, 91):
        three_25_categories[5] = three_25_categories[5] + 1
    elif num in range(91, 106):
        three_25_categories[6] = three_25_categories[6] + 1
    elif num in range(106, 121):
        three_25_categories[7] = three_25_categories[7] + 1
```

```python
one_25_categories
```

```python
three_25_categories
```

```python
labels = ['< 15', 30, 45, 60, 75, 90, 105, '> 120']
```

```python
l = []
l1 = [[labels[i], one_25_categories[i], "Red"] for i in range(len(labels))]
for i in l1:
    l.append(i)
l2 = [[labels[i], three_25_categories[i], "Green"] for i in range(len(labels))]
for i in l2:
    l.append(i)
l1 = [e[0] for e in l]
l2 = [e[1] for e in l]
l3 = [e[2] for e in l]
d1 = pd.DataFrame({'Signals Time (s)': l1, 'Count':l2, 'Color':l3})
```

```python
sns.set_style('white')
plt.figure(figsize=(20,10))
bar_plot = sns.barplot(x = 'Count', y = 'Signals Time (s)', data = d1, orient= 'h', hue = 'Color', palette=["red", "green"])
sns.despine(left=True)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.legend(fontsize='xx-large', title_fontsize='40')
plt.xlabel('Count',fontsize=18, labelpad=10)
plt.ylabel('Signal Time (s)', fontsize=18)
plt.title("Distribution of Road (RSE25 Major)", fontsize = 20, loc = 'center', pad = 20)
plt.savefig("Distribution of Road (RSE25 Major).jpg", bbox_inches='tight')
```
