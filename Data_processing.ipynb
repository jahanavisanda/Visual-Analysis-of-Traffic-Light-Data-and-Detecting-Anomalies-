plotting line plot for a chunk of RSE-25 data

```python
df_line = pd.read_csv("data/RSE25_rawfixed.csv")
```

```python
timelist1 = [datetime.fromtimestamp(l).strftime('%m-%d:%H')+":00" for l in df_line['Timestamp'].tolist()]
```

```python
df_line.insert(1, 'Time', timelist1)
```

```python
fig = plt.figure()
axes = df_line[['Time','CurrentIndication_MajorStreet','Duration_Indication_MajorStreet']].plot(subplots=True, figsize=(20,8), x = 'Time', legend = True)
axes[0].legend(loc='upper right', bbox_to_anchor=(1.004, 1.15))
axes[1].set_ylabel("Signal Time (s)")
axes[0].set_ylabel("Indicator State")
plt.suptitle("Major Street Indicator State and Duration (RSE25)", fontsize = 15)
plt.savefig("Major Road State and Duration Line Plot (RSE25).jpg", bbox_inches='tight')
```

```python
fig = plt.figure()
axes = df_line[['Time','CurrentIndication_MinorStreet','Duration_Indication_MinorStreet']].plot(subplots=True, figsize=(20,8), x = 'Time', legend = True)
axes[0].legend(loc='upper right', bbox_to_anchor=(1.004, 1.15))
axes[1].set_ylabel("Signal Time (s)")
axes[0].set_ylabel("Indicator State")
plt.suptitle("Minor Street Indicator State and Duration (RSE25)", fontsize = 15)
plt.savefig("Minor Road State and Duration Line Plot.jpg", bbox_inches='tight')
```
